```python
flor = Image.open("rose.tif")
flor_w, flor_h = flor.size
print(flor_w, "x", flor_h)
displayImage(flor, flor_w, flor_h)
```

```python
flor_512 = downscaleImage(flor, 512, 512)
flor_256 = downscaleImage(flor, 256, 256)
flor_128 = downscaleImage(flor, 128, 128)
flor_64 = downscaleImage(flor, 64, 64)
flor_32 = downscaleImage(flor, 32, 32)

flor_grid_u = np.concatenate((np.asarray(flor), np.asarray(flor_512.resize((flor_w, flor_h), resample=Image.Resampling.NEAREST))), axis=1)
flor_grid_u = np.concatenate((np.asarray(flor_grid_u), np.asarray(flor_256.resize((flor_w, flor_h), resample=Image.Resampling.NEAREST))), axis=1)

flor_grid_l = np.concatenate((np.asarray(flor_128.resize((flor_w, flor_h), resample=Image.Resampling.NEAREST)), np.asarray(flor_64.resize((flor_w, flor_h), resample=Image.Resampling.NEAREST))), axis=1)
flor_grid_l = np.concatenate((np.asarray(flor_grid_l), np.asarray(flor_32.resize((flor_w, flor_h), resample=Image.Resampling.NEAREST))), axis=1)

flor_grid = np.concatenate((flor_grid_u, flor_grid_l))
```

```python
displayImage(Image.fromarray(flor_grid), 1024*3, 1024*2)
```
