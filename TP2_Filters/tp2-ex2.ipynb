## Load and display Barbara image

```python
b = cv2.imread('img/barbara.jpg')
h, w, channels = b.shape
plotImage(b,w,h)
```

## imagen de barbara filtrada por h0

\begin{equation}
    h_0 = 
    \begin{bmatrix}
    0 & 1/6 & 0 \\
    1/6 & 1/3 & 1/6 \\
    0 & 1/6 & 0 
    \end{bmatrix}
\end{equation}

```python
h0 = np.array([[0,1/6,0],[1/6,1/3,1/6],[0,1/6,0]])
img_f = cv2.filter2D(b,-1,h0)
plotImage(img_f,w,h)
```

## Imagen de barbara filtrada por un pasabajos

```python
h1 = disk(5) #Low pass
img_f = cv2.filter2D(b,-1,h1)
plotImage(img_f,w,h)
```

## Imagen de Barbara filtrada por un pasa altos

```python
h2 = unsharp(0.2) #High pass
img_f = cv2.filter2D(b,-1,h2)
plotImage(img_f,w,h)
```
