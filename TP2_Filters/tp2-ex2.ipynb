```python
b = Image.open('img/barbara.jpg')
w,h = b.size
displayImage(b, w, h)
```

```python
h0 = np.array([[0,1/6,0],[1/6,1/3,1/6],[0,1/6,0]])
img_f = scipy.ndimage.correlate(b, h0, mode='nearest')
fig = px.imshow(img_f,binary_string=True)
fig.update_layout(width=w, height=h, margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

```python
h1 = skimage.morphology.disk(radius=0.9999)    #Low pass
img_f = scipy.ndimage.correlate(b, h1, mode='nearest')
fig = px.imshow(img_f,binary_string=True)
fig.update_layout(width=w, height=h, margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

```python
h2 = unsharp(0.2) #High pass
img_f = scipy.ndimage.correlate(b, h2, mode='nearest')
fig = px.imshow(img_f,binary_string=True)
fig.update_layout(width=w, height=h, margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
